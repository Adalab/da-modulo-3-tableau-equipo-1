# Agrupación job_title a job_area
    - para mejor visualización y entendimento agrupamos algunos puestos en areas de trabajo.

```python
df.job_title.unique()
```

```python
# FUNCION AGRUPAR

                     
def job_area(column):
    
    patron_1 = 'Data Scien\w+'                              
    patron_2 = 'Data Engineer'                              
    patron_3 = 'M\w+ L\w+'                   
    patron_4 = 'Data Analy\w+'                                
    patron_5 = 'Research Scientist'   
    
    if re.findall(patron_1, column):
        return "Data Scientist"
    elif re.findall(patron_2, column):
        return "Data Engineer"
    elif re.findall(patron_3, column):
        return "Machine Learning Engineer"
    elif re.findall(patron_4, column):
        return "Data Analyst"
    elif re.findall(patron_5, column):
        return "Research Scientist"
    else:
       return column

   

```

```python
# añadimos nueva columna con nuevas categorias al data

df['job_area'] = (df['job_title'].apply(job_area))
```
